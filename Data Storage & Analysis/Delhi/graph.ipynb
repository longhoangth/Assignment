```python
import datetime, warnings, scipy 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.rcParams["patch.force_edgecolor"] = True
plt.style.use('fivethirtyeight')
mpl.rc('patch', edgecolor='dimgray', linewidth=1)

from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = "last_expr"
pd.options.display.max_columns = 50
%matplotlib inline
warnings.filterwarnings("ignore")
import math
```

```python
df = pd.read_csv('india-air-quality-2015-2020/Delhi/aqi.csv', parse_dates=["Datetime"])
```

```python
df.head()
```

```python
df.plot(x='Datetime',y='AQI_calculated',figsize=(20,8))
plt.ylabel('AQI')
plt.xlabel('Time')
plt.xticks(rotation=45)

plt.show()
```

```python
df['Date']=df["Datetime"].dt.date
df['Month']=df["Datetime"].dt.month
df['Year']=df["Datetime"].dt.year
df['Hour'] = df['Datetime'].dt.hour
df['Day'] = df['Datetime'].dt.day
```

```python
df
```

```python
df.groupby('Day')['AQI_calculated'].mean().plot()
plt.ylabel('AQI')
plt.xlabel('Day')
plt.xticks(rotation=45)
plt.title('Daily average AQI')
plt.show()
```

```python
df_day=df.loc[:,['Day','AQI_calculated']]
sns.boxplot(x='Day', y='AQI_calculated',data=df_day)
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby('Month')['AQI_calculated'].mean().plot()
plt.ylabel('AQI')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.title('Monthly average AQI')
plt.show()
```

```python
df_month=df.loc[:,['Month','AQI_calculated']]
sns.boxplot(x='Month', y='AQI_calculated',data=df_month)
```

```python
df.groupby('Year')['AQI_calculated'].mean().plot()
plt.ylabel('AQI')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.title('Yearly average AQI')
plt.show()
```

```python
df_year=df.loc[:,['Year','AQI_calculated']]
sns.boxplot(x='Year', y='AQI_calculated',data=df_year)
```

```python
df.groupby('Hour')['AQI_calculated'].mean().plot()
plt.ylabel('AQI')
plt.xlabel('Hour')
plt.xticks(rotation=45)
plt.title('Hourly average AQI')
plt.show()
```

```python
df_hour=df.loc[:,['Hour','AQI_calculated']]
sns.boxplot(x='Hour', y='AQI_calculated',data=df_hour)
```
